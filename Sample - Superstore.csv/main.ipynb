```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Load The Data
df=pd.read_csv("Sample - Superstore.csv", encoding='ISO-8859-1') ## ENCOADING IS USE TO HANDLE SPECIAL CHARCTER
```

```python
print("First 5 Rows : ")
df.head()
```

```python
print("Last 5 Rows : ")
df.tail()
```

```python
## data analysis
print("Total Rows and cols : ",df.shape)
print("Total Rows: ",df.shape[0])
print("Total Cols: ",df.shape[1])
print("Information : ",df.info())
```

```python
print("Describe : ",df.describe())
```

```python
print("Cols Name : ",df.columns)
```

```python
## Cleaning The Data
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
## No dulplicated in dataset
df.duplicated().sum()
```

```python

df.columns
```

```python
##Total Sales, Profit, and Quantity Overview
print("Total Sales, Profit, and Quantity Overview : ")
print("Total Sales : ",df["Sales"].sum())
print("Total Profit : ",df["Profit"].sum())
print("Total Quantity  :  ",df["Quantity"].sum())
```

```python
##sales vs profit by category

category_data=df.groupby("Category")[["Sales","Profit"]].sum()
category_data
```

```python
category_data.plot(kind="bar")
plt.title('Sales vs Profit by Category')
plt.ylabel('Amount')
plt.xlabel('Category')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
```

```python
##Sales vs Profit by Sub-Category
sub_Category_data=df.groupby("Sub-Category")[["Sales","Profit"]].sum()
sub_Category_data
```

```python
sub_Category_data.plot(kind="bar")
plt.title("Sales vs Profit by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout
```

```python
##Region-wise Profit
region_profit=df.groupby("Region")[["Sales","Profit"]].sum()
region_profit
```

```python
region_profit.plot(kind='bar')
plt.title("Region-wise Profit")
plt.xlabel("Region")
plt.ylabel("Amount")
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
## Ship Mode Usage
ship_mode=df["Ship Mode"].value_counts()
ship_mode
```

```python
plt.pie(
    ship_mode,
    labels=ship_mode.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title("Ship Mode Usage")
```

```python
## Top 10 Customers by sales
top_customers=df.groupby("Customer Name")[["Sales"]].sum()
TOP_Customer=top_customers.sort_values(by="Sales",ascending=False).head(10)
TOP_Customer
```

```python
TOP_Customer.plot(kind="bar",color="orange")
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.tight_layout()
plt.grid(True)

plt.title("Top 10 Customers by sales")
```

```python
## Top 10 States by Profit
top_states=df.groupby("State")[["State","Profit"]].sum()
top_states_profit=top_states.sort_values(by="Profit",ascending=False).head(10)
top_states_profit
```

```python
top_states_profit.plot(kind="bar",color="green")
plt.title("Top 10 Cities by Profit")
plt.xlabel("City")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
##Top 5 Products by Sales

top_products=df.groupby("Product Name")[["Sales"]].sum()
top_5_product=top_products.sort_values(by="Sales",ascending=False).head(5)
top_5_product
```

```python
plt.pie(
    top_5_product["Sales"],
    labels=top_5_product.index,
    autopct='%1.1f%%',
    shadow=True
)
```

```python
##Top 10 Profitable Products
top_profit_product=df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)
top_profit_product
```

```python
top_profit_product.plot(kind="bar",color="green",figsize=(12,6),edgecolor="black")
plt.title("Top 10 Profitable Products")
plt.xlabel("Product Name")
plt.ylabel("Profit Amount")
plt.grid(True)
plt.tight_layout()
```

```python

## Loss-Making Products 
loss_product=df.groupby("Product Name")[["Profit"]].sum().sort_values(by="Profit").head(10)
loss_product
```

```python
loss_product.plot(kind="bar", figsize=(12,6), label="Lo", color="red",title="Top 10 Loss-Making Products")
plt.ylabel("Profit (Negative = Loss)") 
plt.legend("loss")
plt.show()
```

```python
## Monthly Sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"]=df["Order Date"].dt.to_period("M")

monthly_sales=df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line",marker='o', color='blue')

plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()

```
